```python
import pandas as pd
import matplotlib.pyplot as plt

# Load processed dataset
df = pd.read_csv("../data/processed/naukri_data_science_jobs_india_processed.csv")

# Basic information
print(f"Total Rows    : {df.shape[0]}")
print(f"Total Columns : {df.shape[1]}")

df.info()
```

```python
df.describe(include="all").T
```

```python
# 1. Top Hiring Companies
```

```python
top_companies = (
    df["company"]
    .value_counts()
    .head(10)
)

top_companies
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

top_companies.sort_values().plot(kind="barh")

plt.title("Top 10 Hiring Companies")
plt.xlabel("Number of Job Postings")
plt.ylabel("Company")

plt.tight_layout()
plt.show()
```

```python
# 2. Top Hiring Locations
```

```python
top_locations = (
    df["primary_location"]
    .value_counts()
    .head(10)
)

top_locations
```

```python
plt.figure(figsize=(10,6))

top_locations.sort_values().plot(kind="barh")

plt.title("Top 10 Hiring Locations")
plt.xlabel("Number of Job Postings")
plt.ylabel("Location")

plt.tight_layout()
plt.show()
```

```python
# 3. Most In-Demand Data Roles
```

```python
# Top job roles
top_roles = (
    df["role_category"]
    .value_counts()
    .head(10)
)

top_roles
```

```python
plt.figure(figsize=(10,6))

top_roles.sort_values().plot(kind="barh")

plt.title("Top 10 Most In-Demand Data Roles")
plt.xlabel("Number of Job Postings")
plt.ylabel("Role")

plt.tight_layout()
plt.show()
```

```python
# 4. Salary Distribution
```

```python
# Salary distribution
salary_distribution = (
    df["salary_category"]
    .value_counts()
    .sort_index()
)

salary_distribution
```

```python
plt.figure(figsize=(8,5))

salary_distribution.plot(kind="bar")

plt.title("Job Distribution by Salary Category")
plt.xlabel("Salary Category")
plt.ylabel("Number of Job Postings")

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 5. Experience Level Distribution
```

```python
experience_distribution = (
    df["experience_level"]
    .value_counts()
)

experience_distribution
```

```python
plt.figure(figsize=(8,5))

experience_distribution.plot(kind="bar")

plt.title("Job Distribution by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Number of Job Postings")

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
